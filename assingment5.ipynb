```python
from transformers import pipeline
```

#1. Story writer

```python
story_writer_pipeline = pipeline("text-generation", model="gpt2-medium")
```

**Try out different Prompts below**

```python
prompt = "I saw him hiding"
```

```python
generated_text = story_writer_pipeline(prompt, max_length=100, do_sample=True)
```

```python
print(generated_text[0]['generated_text'])
```

#2.sentiment analysis

```python
sentiment_pipeline = pipeline("sentiment-analysis", model="bert-base-uncased")
```

```python
text = "it is very sad to hear that he would do such a thing"
```

```python
result = sentiment_pipeline(text)
```

```python
print("Sentiment:", result[0]['label'])
print("Score:", result[0]['score'])
```

3.text summarizer

```python
summarization_pipeline = pipeline("summarization", model="t5-small")
```

**Enter the Text to be Summarized:**

```python
article = """
In a quaint village, Lily found a hidden door in an ancient oak tree. Inside, a magical world awaited, filled with blooming flowers and whispering trees.
"""
```

```python
summary = summarization_pipeline(article, max_length=200, min_length=50, do_sample=False)
```

```python
print(summary[0]['summary_text'])
```
